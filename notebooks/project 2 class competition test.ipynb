```python
import joblib
import pandas as pd
from pathlib import Path
```

```python
PROJECT_ROOT = Path.cwd().parent

DATA_DIR = PROJECT_ROOT / "data"
OUTPUT_DIR = PROJECT_ROOT / "output"

OUTPUT_DIR.mkdir(exist_ok=True)
```

```python
#loading the saved best model XGBoost
xgb_best_model=joblib.load(OUTPUT_DIR / 'xgboost_best_model.joblib')
print("model loaded")
```

```python
#loading the preprocessed competition test data
test_data=pd.read_csv(OUTPUT_DIR / "final_test_data_selected.csv")
X_competition=test_data  
```

```python
#using the loaded best model to predict the labels
y_pred_competition =xgb_best_model.predict(X_competition)
#converting numeric predictions 0/1 to Benign/Trojan
y_pred_class=['Benign' if pred==0 else 'Trojan' for pred in y_pred_competition]
#dataFrame with the predicted labels
competition_predictions=pd.DataFrame({'Class': y_pred_class})
```

```python
competition_predictions.to_csv(OUTPUT_DIR / "xgb_competition_predictions.csv", index=False)
print("predictions saved as csv")
```
